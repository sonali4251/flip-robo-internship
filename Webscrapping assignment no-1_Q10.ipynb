```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from lxml import html
```

```python
page= requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en&vq=en')
page
```

```python
soup=BeautifulSoup(page.content)
rank=soup.find('td',class_="gsc_mvt_p")
rank
```

```python
rank.text
```

```python
Rank=[]
for i in soup.find_all('td',class_="gsc_mvt_p"):
    Rank.append(i.text)
Rank
```

```python
publication=soup.find('td',class_="gsc_mvt_t")
rank
Publications=[]
for i in soup.find_all('td',class_="gsc_mvt_t"):
    Publications.append(i.text)
Publications
```

```python
h5index=soup.find('a',class_="gs_ibl gsc_mp_anchor")
h5index
H5index=[]
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    H5index.append(i.text)
H5index
```

```python
h5median=soup.find('span',class_="gs_ibl gsc_mp_anchor")
h5median
H5median=[]
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    H5median.append(i.text)
H5median
```
