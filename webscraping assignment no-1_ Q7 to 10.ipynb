# Write a python program to scrape the details of most downloaded articles from AI in last 90 days

```python
page= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
soup=BeautifulSoup(page.content)
title=soup.find('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg")
title.text
Article_title=[]
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Article_title.append(i.text)
Article_title

Author=soup.find('span',class_="sc-1w3fpd7-0 dnCnAO")
Author.text
Author_Name=[]
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    Author_Name.append(i.text)
Author_Name

Date=soup.find('span',class_="sc-1thf9ly-2 dvggWt")
Date.text
Publication_date=[]
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    Publication_date.append(i.text)
Publication_date

url=soup.find('a',class_="sc-5smygv-0 fIXTHm")
url.get('data-src')
Paper_url=[]
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    Paper_url.append(i.get('href'))
Paper_url

df=pd.DataFrame({'Article_title':Article_title,'Author_Name':Author_Name,'Publication_date':Publication_date,'Paper_url':Paper_url})
df
```

# Write a python program to scrape mentioned details from dineout.co.in¶

```python
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
soup=BeautifulSoup(page.content)
restaurant_name=soup.find('a',class_="restnt-name ellipsis")
restaurant_name.text
Restaurant_Name=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Restaurant_Name.append(i.text)
Restaurant_Name
cuisine=soup.find('ul',class_="options")
cuisine
Cuisine=[]
for i in soup.find_all('ul',class_="options"):
    Cuisine.append(i.text)
Cuisine
location=soup.find('div',class_="restnt-loc ellipsis")
location.text
Location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
Location

price=soup.find('span',class_="double-line-ellipsis")
price.text.split('|')
price.text.split('|')[0]
Price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    Price.append(i.text.split('|')[0])
Price
ratings=soup.find('div',class_="restnt-rating rating-4")
Ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
Ratings

image=soup.find('img',class_="no-img")
image.get('data-src')

Images=[]
for i in soup.find_all('img',class_="no-img"):
    Images.append(i.get('data-src'))
Images

print(len(Restaurant_Name),len(Cuisine),len(Location), len(Price),len(Ratings),len(Images))

df=pd.DataFrame({'Restaurant_Name':Restaurant_Name,'Location':Location,'Price':Price,'Ratings':Ratings,'Images':Images})
df
```

```python
page= requests.get('https://www.cnbc.com/world/?region=world')
page
soup=BeautifulSoup(page.content)
headline=soup.find('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail")
Headline=[]
for i in soup.find_all('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"):
    Headline.append(i.text)
Headline
```

```python
time=soup.find('span',class_="RiverByline-datePublished")
time=[]
for i in soup.find_all('span',class_="RiverByline-datePublished"):
    time.append(i.text)
time
```

```python
# to find the nwes links
for links in soup.find_all('a'):
    link =links.get('href')
    if link[0:3]=='../' and link !='#':
        print("https://www.cnbc.com/world/?region=world"+ link[2:len(link)])
    elif link[0]=="/" and link !='#':
        print ('https://www.cnbc.com/world/?region=world'+ link)
```
